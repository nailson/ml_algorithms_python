## Test Least Squares Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sys.path.append('..')
from simple_least_squares_regression import LeastSquaresRegression
```

```python
X = 2 * np.random.rand(100) 
y = 4 + 3 * X + np.random.randn(100)

data = pd.DataFrame({'covariate':X, 'response':y})

plt.scatter(X, y)
```

```python
lr_model = LeastSquaresRegression()
lr_model.fit(data.covariate, data.response)
```

```python
print("Intercept: {}\nSlope: {}".format(lr_model.intercept, lr_model.slope))
```

```python
y_pred = lr_model.predict(X)

plt.scatter(X,y)
plt.plot(X, y_pred, color='red')
```
